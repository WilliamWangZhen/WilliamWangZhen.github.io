### Medium complexity - Birthday Paradox

Birthday paradox concerns the probability that, in a set of `n` randomly chosen people, some pair of them will have the same birthday. You can read about this problem at https://en.wikipedia.org/wiki/Birthday_problem.

Write a function called **`has_duplicates()`**, that takes a list and returns `True` if there is any element that appears more than once. It should not modify the original list.

If there are 45 students in the class, what are the chances that any two students will have the same birthday? You can estimate this probability by generating random samples of 45 birthdays and checking for matches.

**Hints:**

- probability

```python
def has_duplicates(num_students, num_simulations):
    total_duplicates = 0
    for i in range(num_simulations):
        # generate birthday list
        l = generateBirthdayList(num_students)
        # check duplicates, output is a list 
        l1 = check_duplicates(l)

        # output:
        # 1. number of students
        # 2. number of iterations/samples
        # 3. list of duplicate days for each iteration, where duplicates are found
        # 4. probability
        print("Simulation #" + str(i + 1))
        print("number of students: " + str(num_students))
        print("Samples: \n" + str(l))
        print("Duplicate days: " + str(l1))
        print("numbers of duplicate days: " + str(len(l1)))
        print("Probability " + str(round(len(l1)/num_students * 100,2)) + "%")
        print("=" * 40)

                
def generateBirthdayList(num_students):
    '''(num) -> bolean
    return a list contains "num_students" random numbers between 1 and 365 inclusive
    sort numbers asending
    '''
    import random
    birthday_List = [random.randint(1,366) for i in range(0, num_students)]
    birthday_List.sort()
    return birthday_List


def check_duplicates(birthday_List):
    dict1 = {}
    for num in birthday_List:
        if birthday_List.count(num) > 1:
            if num in dict1:
                dict1[num] += 1
            else:
                dict1[num] = 1
    keys_List = []
    for key in dict1:
        keys_List.append(key)
    return keys_List
```
